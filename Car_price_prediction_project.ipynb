**DATA PREPROCESSING**

```python
df['registered_year'].value_counts()
```

```python
df['registered_year']=df['registered_year'].str.split().str[-1]    # for extracting the year
df['registered_year']=pd.to_numeric(df['registered_year'])        # for converting into numerical value
```

```python

df['resale_price']=df['resale_price'].str.replace('₹','').str.replace(',','').str.replace('null','0')
df.dtypes
```

```python
df['resale_price'] = df['resale_price'].apply(lambda x: float(x.split()[0]) * 10000000 if 'Crore' in x else (float(x.split()[0]) * 100000 if 'Lakh' in x else float(x)))
```

```python
# for changing the string element from columns-engine_capacity,kms_driven,max_power,mileage

df['engine_capacity']=df['engine_capacity'].str.replace('cc','')
df['kms_driven']=df['kms_driven'].str.replace('Kms','')
df['kms_driven'] = df['kms_driven'].str.replace(',', '')   # for changing commas in kms_driven

df['max_power']=df['max_power'].str.replace('bhp','')
df['max_power']=df['max_power'].str.replace('PS','')
df['max_power']=df['max_power'].str.replace('KW','')

df['mileage']=df['mileage'].str.replace('kmpl','')
df['mileage']=df['mileage'].str.replace('km/kg','')
```

```python
# null replaced with 0
# converted into float

df['engine_capacity']=df['engine_capacity'].str.replace('null','0').astype(float)
df['kms_driven']=df['kms_driven'].str.replace('null','0').astype(float)
df['max_power']=df['max_power'].str.replace('null','0')
df['max_power'] = df['max_power'].str.extract(r'(\d+(?:\.\d+)?)').astype(float) #The function used to extract numeric values
df['mileage']=df['mileage'].str.replace('null','0').astype(float)
```

```python
df['registered_year'].fillna(df['registered_year'].mean(),inplace=True)
df['engine_capacity'].fillna(df['engine_capacity'].mean(),inplace=True)
df['insurance'].fillna(df['insurance'].mode()[0],inplace=True)
df['kms_driven'].fillna(df['kms_driven'].mode()[0],inplace=True)
df['owner_type'].fillna(df['owner_type'].mode()[0],inplace=True)
df['max_power'].fillna(df['max_power'].mean(),inplace=True)
df['seats'].fillna(df['seats'].median(),inplace=True)
df['mileage'].fillna(df['mileage'].mean(),inplace=True)
```

**Categorical Variable Encoding - Get dummies**

```python
df1=pd.get_dummies(df[['insurance','transmission_type','owner_type','fuel_type','body_type','city']],drop_first=True)
df1
```

```python
df1.shape
```

```python
df2=pd.concat([df,df1],axis=1)
df2
```

Removing unwanted columns

```python
df2.drop(['Unnamed: 0','full_name','insurance','transmission_type','owner_type','fuel_type','body_type','city'],axis=1,inplace=True)
df2
```

**FEATURE SELECTION**

```python
correlation_matrix=df2.corr()
correlation_matrix
```

```python
correlation_with_output=correlation_matrix['resale_price'].drop('resale_price')
correlation_with_output
```

```python
most_related_columns = correlation_with_output.abs().sort_values(ascending=False)
most_related_columns=pd.DataFrame(most_related_columns.items(),columns=['Feature','Correlation'])
most_related_columns
```

```python

# Plotting the correlation of each feature with 'resale_price' using a horizontal bar plot

import matplotlib.pyplot as plt
plt.figure(figsize=(10,20))
plt.barh(most_related_columns['Feature'], most_related_columns['Correlation'], color='g')
plt.xlabel('Correlation with Resale Price',size=12)
plt.ylabel('Feature',size=12)
plt.title('CORRELATION OF FEATURES WITH RESALE PRICE')
plt.show()
```

```python
# columns having  low correlation
threshold=.02
# Filter the correlation matrix to include only columns with correlation less than the threshold
low_corr_columns= most_related_columns[most_related_columns['Correlation'].abs() < threshold]
print("Columns with correlation less than 0.02:", low_corr_columns)
```

```python
low_corr_columns_drop=low_corr_columns['Feature']
df2=df2.drop(columns=low_corr_columns_drop)
df2
```

```python
y=df2.iloc[:,0]
y
```

```python
x=df2.iloc[:,1:]
x
```

Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
x_train
```

**Model initialization**


DecisionTree


```python

from sklearn.tree import DecisionTreeRegressor
model1=DecisionTreeRegressor(criterion='friedman_mse')
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
y_pred
```

```python
error_decisionTree=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
error_decisionTree
```

**Performance of Model**

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("PERFORMANCE OF DECISIONTREE")
print(" ")
print("Mean absolute error is ",mean_absolute_error(y_test,y_pred))
print("Mean_absolute_percentage_error is",mean_absolute_percentage_error(y_test,y_pred))
print("Mean squared error is ",mean_squared_error(y_test,y_pred))
print('r2_score is',r2_score(y_test,y_pred))
```

Linear Regression

```python

from sklearn.linear_model import LinearRegression
model2=LinearRegression()
model2.fit(x_train,y_train)
y_pred_linear=model2.predict(x_test)
y_pred_linear
```

```python
error_linear=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred_linear,'Difference':y_test-y_pred_linear})
error_linear
```

```python
# SCATTER PLOT

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(error_linear['Actual_value'], error_linear['Difference'], color='blue', alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')  # Add horizontal line at y=0
plt.xlabel('Actual Value',size=15)
plt.ylabel('Difference',size=15)
plt.title('Actual Value vs. Difference (Linear Regression)',size=20)
plt.grid(color='g',linewidth=.5,linestyle=':')
plt.show()
```

```python
print("PERFORMANCE OF LINEAR REGRESSION MODEL")
print(" ")
print("Mean absolute error is ",mean_absolute_error(y_test,y_pred_linear))
print("Mean_absolute_percentage_error is",mean_absolute_percentage_error(y_test,y_pred_linear))
print("Mean squared error is ",mean_squared_error(y_test,y_pred_linear))
print('r2_score is',r2_score(y_test,y_pred_linear))
```

```python
print('Slope is',model2.coef_)
print('Constant is',model2.intercept_)
```
